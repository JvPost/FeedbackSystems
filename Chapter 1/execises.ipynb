## Exercise 1.5

1.5 (Integral action) We say that a system with a constant input reaches steady state if the output of the system approaches a constant value as time increases. Show that a controller with integral action, such as those given in equations (1.4) and (1.5), gives zero error if the closed loop system reaches steady state.

# Why Integral Action Implies Zero Steady-State Error (If System Reaches Steady State)

We are given:

- A **controller with integral action**:
  $$
  u(t) = K_p e(t) + K_i \int_0^t e(\tau)\, d\tau + K_d \frac{de(t)}{dt}
  $$

- A **constant reference** input: $r(t) = r$
- The **closed-loop system reaches steady state**, meaning:
  $$
  \lim_{t \to \infty} y(t) = y_\infty \quad \text{and} \quad \lim_{t \to \infty} \dot{y}(t) = 0
  $$

Define the tracking error:
$$
e(t) = r - y(t)
$$

---

## Step-by-step argument:

### 1. Assume steady state is reached

Then $y(t)$ converges to a constant value.  
Therefore, $e(t) \to e_\infty$ as $t \to \infty$.

Also, the control input $u(t)$ must converge to a constant:
$$
\lim_{t \to \infty} u(t) = u_\infty
$$

---

### 2. Suppose (for contradiction) that $e_\infty \ne 0$

Then for large $t$, $e(t)$ stays near $e_\infty$.  
So the integral term grows approximately linearly:
$$
\int_0^t e(\tau)\, d\tau \approx e_\infty \cdot t \to \infty
$$

Hence:
$$
u(t) = K_p e(t) + K_i \int_0^t e(\tau)\, d\tau + K_d \frac{de(t)}{dt}
\to \infty
$$

---

### 3. But this contradicts the steady-state assumption

If $u(t)$ grows unbounded, and the system is causal and physically reasonable (e.g. $\dot{y}(t) = u(t)$ or similar), then $y(t)$ cannot remain constant.

This violates the assumption that the system reaches steady state.

---

Therefore, our assumption must be wrong:  
$$
\lim_{t \to \infty} e(t) = 0
$$

So:  
**If a system with integral action reaches steady state, then the steady-state error must be zero.**
